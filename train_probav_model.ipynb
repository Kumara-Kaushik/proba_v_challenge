```python
# Importing useful libraries
import fastai
from fastai.basics import *
from fastai.vision import *
from fastai.callbacks import *
from fastai.utils.mem import *

from torchvision.models import vgg16_bn
```

```python
# Defining training data path
path_hr = proba_data.ls()[1]/'HR_imgs'
path_lr = proba_data.ls()[1]/'LR_imgs'
```

```python
# Defining batch size and model
bs,size=32, 128
arch = models.resnet34
```

```python
# Creating Image list
src = ImageImageList.from_folder(path_lr).split_by_rand_pct(0.2, seed=42)
```

```python
def get_data(bs,size):
    data = (src.label_from_func(lambda x: path_hr/x.name)
           .transform(get_transforms(max_zoom=2.), size=size, tfm_y=True)
           .databunch(bs=bs).normalize(imagenet_stats, do_y=True))

    data.c = 3
    return data
```

```python
# Creating training batches
data = get_data(bs,size)
```

```python
# Displaying a training batch
data.show_batch(ds_type=DatasetType.Valid, rows=2, figsize=(9,9))
```

## Feature loss

```python
t = data.valid_ds[0][1].data
t = torch.stack([t,t])
```

```python
t.size()
```

```python
def gram_matrix(x):
    n,c,h,w = x.size()
    x = x.view(n, c, -1)
    return (x @ x.transpose(1,2))/(c*h*w)
```

```python
gram_matrix(t)
```

```python
# Defining base loss
base_loss = F.l1_loss
```

```python
# Grabbing vgg model
vgg_m = vgg16_bn(True).features.cuda().eval()
requires_grad(vgg_m, False)
```

```python
# Taking out the required layers from the vgg model to buid the loss function
blocks = [i-1 for i,o in enumerate(children(vgg_m)) if isinstance(o,nn.MaxPool2d)]
blocks, [vgg_m[i] for i in blocks]
```

```python
# building the perception loss function hook from vgg model for fastai training
class FeatureLoss(nn.Module):
    def __init__(self, m_feat, layer_ids, layer_wgts):
        super().__init__()
        self.m_feat = m_feat
        self.loss_features = [self.m_feat[i] for i in layer_ids]
        self.hooks = hook_outputs(self.loss_features, detach=False)
        self.wgts = layer_wgts
        self.metric_names = ['pixel',] + [f'feat_{i}' for i in range(len(layer_ids))
              ] + [f'gram_{i}' for i in range(len(layer_ids))]

    def make_features(self, x, clone=False):
        self.m_feat(x)
        return [(o.clone() if clone else o) for o in self.hooks.stored]
    
    def forward(self, input, target):
        out_feat = self.make_features(target, clone=True)
        in_feat = self.make_features(input)
        self.feat_losses = [base_loss(input,target)]
        self.feat_losses += [base_loss(f_in, f_out)*w
                             for f_in, f_out, w in zip(in_feat, out_feat, self.wgts)]
        self.feat_losses += [base_loss(gram_matrix(f_in), gram_matrix(f_out))*w**2 * 5e3
                             for f_in, f_out, w in zip(in_feat, out_feat, self.wgts)]
        self.metrics = dict(zip(self.metric_names, self.feat_losses))
        return sum(self.feat_losses)
    
    def __del__(self): self.hooks.remove()
```

```python
# Initiating the loss function
feat_loss = FeatureLoss(vgg_m, blocks[2:5], [5,15,2])
```

## Train

```python
# Defining the learner
wd = 1e-3
learn = unet_learner(data, arch, wd=wd, loss_func=feat_loss, callback_fns=LossMetrics,
                     blur=True, norm_type=NormType.Weight)
gc.collect();
```

```python
# Finding optimum learning rate using the learning rate finder
learn.lr_find()
learn.recorder.plot()
```

```python
# Defining learning rate
lr = 1e-3
```

```python
# Defining training function
def do_fit(save_name, lrs=slice(lr), pct_start=0.9):
    learn.fit_one_cycle(10, lrs, pct_start=pct_start)
    learn.save(save_name)
    learn.show_results(rows=1, imgsize=5)
```

```python
# First fit
do_fit('model_1', slice(lr*10))
```

```python
# Unfreezing the layers of the model
learn.unfreeze()
```

```python
# 2nd fitting
do_fit('model_2', slice(1e-5,lr))
```

```python
# Decrease the batch size and increase the image size
data = get_data(12,size*2)
```

```python
# Freeze the layers and train only the head 
learn.data = data
learn.freeze()
gc.collect()
```

```python
# load the model from 2nd fitting
learn.load('model_2')
```

```python
# 3rd Fitting
do_fit('model_3')
```

```python
# Unfreeze the layers of the model
learn.unfreeze()
```

```python
# 4th fitting
do_fit('model_4', slice(1e-6,1e-4), pct_start=0.3)
```
